```python
import pandas as pd
```

```python
hip_pop = pd.read_csv('hip hop_2.csv')
jazz = pd.read_csv('Jazz_2.csv')
kpop = pd.read_csv('K-pop_2.csv')
# pop = pd.read_csv('Pop.csv')
RB = pd.read_csv('R&B_2.csv')
rap = pd.read_csv('Rap_2.csv')
rock = pd.read_csv('rock_2.csv')
rhythm_blues = pd.read_csv('Rhythm and Blues_2.csv')
techno = pd.read_csv('Techno_2.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
figure = plt.figure(figsize=(12, 3))
plt.y_lim = (0, 1)
sns.lineplot(data=jazz[["acousticness", "danceability",
             "energy", "instrumentalness", "liveness", "speechiness"]])
# sns.lineplot(data=pop[["acousticness", "danceability", "energy", "liveness", "speechiness"]])
plt.show()
```

```python
dataset = pd.concat([hip_pop, jazz, kpop, RB, rap,
                    rock, rhythm_blues, techno], ignore_index=True)
```

```python
dataset.to_csv('dataset_3.csv', index=False)
```
